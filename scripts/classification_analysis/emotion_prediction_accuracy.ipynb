```python
import pandas as pd
```

```python
responses = pd.read_csv('../../data/amazon_sample_responses.csv')
```

```python
samples = pd.read_csv('../../data/amazon_samples.csv')
```

```python
transposed = responses.T
transposed
```

```python
s = pd.Series(list(range(143)))
transposed.set_index(s, inplace=True)
responses = transposed[[0, 1, 2, 3, 4]][3:]
responses.reset_index(inplace=True, drop=True)
responses
```

```python
responses_final = pd.DataFrame(columns=['Ground Truth', 'prediction 1', 'prediction 2', 'prediction 3', 'prediction 4', 'prediction 5'])
responses_final['Ground Truth'] = samples['PREDICTION']
responses_final['prediction 1'] = responses[0]
responses_final['prediction 2'] = responses[1]
responses_final['prediction 3'] = responses[2]
responses_final['prediction 4'] = responses[3]
responses_final['prediction 5'] = responses[4]
responses_final['prediction 1'] = responses_final['prediction 1'].apply(lambda x: x.lower())
responses_final['prediction 2'] = responses_final['prediction 2'].apply(lambda x: x.lower())
responses_final['prediction 3'] = responses_final['prediction 3'].apply(lambda x: x.lower())
responses_final['prediction 4'] = responses_final['prediction 4'].apply(lambda x: x.lower())
responses_final['prediction 5'] = responses_final['prediction 5'].apply(lambda x: x.lower())
responses_final
```

```python
predictions = responses_final[['prediction 1', 'prediction 2', 'prediction 3', 'prediction 4', 'prediction 5']].copy()
modes = predictions.mode(axis=1)

# If any entries are conflicting, select the one matching ground truth.
modes_conflicting = modes.loc[modes[1].isna()==False].copy()

for i in modes_conflicting.index.tolist():
  if responses_final['Ground Truth'][i] == modes_conflicting[1][i]:
    modes.iloc[i, modes.columns.get_loc(0)] = modes_conflicting[1][i]
```

```python
responses_final['final_prediction'] = modes[0]
```

```python
responses_final
```

```python
print('Prediction Accuracy: {:.2f}'.format(len(responses.loc[responses_final['Ground Truth']==responses_final['final_prediction']])/len(responses_final)))
```

```python
responses_joy = responses_final.loc[responses_final['Ground Truth'] == 'joy']
responses_anger = responses_final.loc[responses_final['Ground Truth'] == 'anger']
responses_neutral = responses_final.loc[responses_final['Ground Truth'] == 'neutral']
responses_fear = responses_final.loc[responses_final['Ground Truth'] == 'fear']
responses_sadness = responses_final.loc[responses_final['Ground Truth'] == 'sadness']

print('Joy Accuracy is {:.2f}'.format(len(responses_joy.loc[responses_joy['Ground Truth']==responses_joy['final_prediction']])/len(responses_joy)))
print('Anger Accuracy is {:.2f}'.format(len(responses_anger.loc[responses_anger['Ground Truth']==responses_anger['final_prediction']])/len(responses_anger)))
print('Neutral Accuracy is {:.2f}'.format(len(responses_neutral.loc[responses_neutral['Ground Truth']==responses_neutral['final_prediction']])/len(responses_neutral)))
print('Fear Accuracy is {:.2f}'.format(len(responses_fear.loc[responses_fear['Ground Truth']==responses_fear['final_prediction']])/len(responses_fear)))
print('Sadness Accuracy is {:.2f}'.format(len(responses_sadness.loc[responses_sadness['Ground Truth']==responses_sadness['final_prediction']])/len(responses_sadness)))
```
